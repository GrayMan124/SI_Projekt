```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data=pd.read_csv("prices-split-adjusted.csv")
```

```python
data.head()
```

```python
data.symbol.unique()
```

```python
apple_data=data[data.symbol=='AAPL']
apple_data=apple_data[apple_data.date>"2014-01-01"]
apple_data=apple_data[apple_data.date<"2016-04-01"]

plt.figure(figsize=(16,8))
plt.plot(apple_data['close'])
```

```python
apple_data.tail()
```

```python
coca_data=data[data.symbol=='KO']
coca_data=coca_data[coca_data.date>"2014-01-01"]
coca_data=coca_data[coca_data.date<"2016-04-01"]

plt.figure(figsize=(16,8))
plt.plot(coca_data['close'])
```

```python
fb_data=data[data.symbol=='FB']
fb_data=fb_data[fb_data.date>"2014-01-01"]
fb_data=fb_data[fb_data.date<"2016-04-01"]

plt.figure(figsize=(16,8))
plt.plot(fb_data['close'])
```

```python
prediction_days= 30
apple_close=apple_data["close"].values.reshape(-1,1)
fb_close=fb_data["close"].values.reshape(-1,1)
coca_close=coca_data["close"].values.reshape(-1,1)
X_train=[]
Y_train=[]

for i in range(prediction_days,len(apple_data)):
    X_train.append(apple_close[i-prediction_days:i,0])     
    Y_train.append(apple_close[i,0])
for i in range(prediction_days,len(fb_data)):
    X_train.append(fb_close[i-prediction_days:i,0])     
    Y_train.append(fb_close[i,0])
for i in range(prediction_days,len(coca_data)):
    X_train.append(coca_close[i-prediction_days:i,0])     
    Y_train.append(coca_close[i,0])
X_train,Y_train=np.array(X_train), np.array(Y_train)
```

```python
X_train.shape
```

```python
from sklearn import model_selection
import scipy.stats as stats
import sklearn.linear_model as lm
from sklearn import datasets, linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
kfold=model_selection.KFold(n_splits=10,shuffle=False)
grid1 = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), linear_model.LinearRegression()),
                    param_grid={'polynomialfeatures__degree': [1,2,3]},
                    cv=kfold,
                    refit=True)

grid1.fit(X_train,Y_train)
grid1.best_params_
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import RandomizedSearchCV
fold=model_selection.KFold(n_splits=5,shuffle=False)
grid2= RandomizedSearchCV(make_pipeline(PolynomialFeatures(degree=2),Lasso()),
                      [{'polynomialfeatures__degree':[1,2,3],'lasso__alpha':[0.3,0.7,1,5,0.5]}],
                     cv=kfold,refit=True,
                          n_jobs=10,
                         pre_dispatch=20)
grid2.fit(X_train,Y_train)
grid2.best_params_
```

```python
from sklearn.linear_model import ElasticNet

grid3= RandomizedSearchCV(make_pipeline(PolynomialFeatures(degree=2),ElasticNet()),
                      [{'polynomialfeatures__degree':[1,2,3],'elasticnet__alpha':[0.1,1,10]}],
                     cv=kfold,refit=True,
                          n_jobs=10,
                         pre_dispatch=20)
grid3.fit(X_train,Y_train)
grid3.best_params_
```

```python
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

grid5= RandomizedSearchCV(RandomForestRegressor(),
                      [{'max_depth':[3,4,5,6],'n_estimators':[1000,100,200,10,50]}],
                     cv=kfold,refit=True,
                          n_jobs=10,
                         pre_dispatch=20)
grid5.fit(X_train,Y_train)
grid5.best_params_
```

```python
google_data=data[data.symbol=='GOOG']
google_data=google_data[google_data.date>'2015-01-01']
plt.figure(figsize=(16,8))
plt.plot(google_data['close'])
```

```python
intel_data=data[data.symbol=='INTC']
intel_data=intel_data[intel_data.date>'2015-01-01']
plt.figure(figsize=(16,8))
plt.plot(intel_data['close'])
```

```python
prediction_days= 30
google_close=google_data["close"].values.reshape(-1,1)
intel_close=intel_data["close"].values.reshape(-1,1)
X_test_google=[]
X_test_intel=[]

for i in range(prediction_days,len(google_data)):
    X_test_google.append(google_close[i-prediction_days:i,0])
for i in range(prediction_days,len(intel_data)):
    X_test_intel.append(intel_close[i-prediction_days:i,0])
X_test_google,X_test_intel=np.array(X_test_google),np.array(X_test_intel)
```

```python
models=[]

models.append(('LR',grid1.best_estimator_))
models.append(('Lasso',grid2.best_estimator_))
models.append(('ElasticNet',grid3.best_estimator_))
models.append(('RandomForest',grid5.best_estimator_))

for name,model in models:
    predicted_google=model.predict(X_test_google)
    predicted_intel=model.predict(X_test_intel)
    plt.figure(figsize=(16,8))
    plt.plot(google_close[prediction_days:130],label='Google Actual Data',color='blue')
    plt.plot(predicted_google[0:100],label=name+'Predicted Google',color='red')
    plt.title("Google")
    plt.show
    plt.figure(figsize=(16,8))
    plt.title("Intel")
    plt.plot(intel_close[prediction_days:130],label='Intel Actual Data',color='black')
    plt.plot(predicted_intel[0:100],label=name+'Predicted Intel',color='orange')
    plt.show
    
    
    
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
data1=apple_data
data2=fb_data
data3=coca_data
scaled_data_apple = scaler.fit_transform(data1['close'].values.reshape(-1,1))
scaled_data_fb = scaler.fit_transform(data2['close'].values.reshape(-1,1))
scaled_data_coca = scaler.fit_transform(data3['close'].values.reshape(-1,1))


prediction_days=30
x_train=[]
y_train=[]
for x in range(prediction_days,len(scaled_data_apple)-4):
    x_train.append(scaled_data_apple[x-prediction_days:x,0])
    y_train.append(scaled_data_apple[x:x+4,0])
for x in range(prediction_days,len(scaled_data_fb)-4):
    x_train.append(scaled_data_fb[x-prediction_days:x,0])
    y_train.append(scaled_data_fb[x:x+4,0])
for x in range(prediction_days,len(scaled_data_coca)-4):
    x_train.append(scaled_data_coca[x-prediction_days:x,0])
    y_train.append(scaled_data_coca[x:x+4,0])
    
    
x_train=np.array(x_train)
y_train=np.array(y_train)
x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
```

```python
print(y_train[2])
```

```python
model=Sequential()

model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=25, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=10))
model.add(Dropout(0.2))
model.add(Dense(units=4))

model.compile(optimizer='adam',loss='mean_squared_error')
model.summary()
```

```python
model.fit(x_train,y_train,epochs=25,batch_size=32)
```
