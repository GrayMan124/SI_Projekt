```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

# Wczytanie danych

```python
data=pd.read_csv("prices-split-adjusted.csv")
data.head()
```

Wybieram tylko akcje The Coca Cola Company

```python
coca=data[data.symbol=='KO']
coca=coca[coca.date>"2013-01-01"]
coca=coca[coca.date<"2016-04-01"]
coca['date']=pd.to_datetime(coca.date)
coca.set_index('date',inplace=True)


plt.figure(figsize=(16,8))
plt.plot(coca['close'])
```

# Przetwarzanie danych 

* Usunicie brakujcych elementw

```python
coca.dropna(inplace=True)
```

Uywam biblioteki ta ( https://technical-analysis-library-in-python.readthedocs.io/en/latest/ ), aby doda wskaniki z analizy technicznej giedy. Wskaniki maj pomc przy przewidywaniu 
cen giedy, bez ciko mi byo stworzy model ktry rzeczywicie przewyidywa ceny akcji

```python
from ta import volume
from ta import trend
from ta import volatility
from ta import momentum
```

* Dodawanie wskanikw z analizy technicznej

```python
accumulation= volume.AccDistIndexIndicator(close=coca['close'],high=coca['high'],low=coca['low'],volume=coca['volume'])
avarge_directional_index=trend.ADXIndicator(close=coca['close'],high=coca['high'],low=coca['low'])
moving_avarge_covergence_divergance=trend.MACD(close=coca['close'])
relative_strength_index=momentum.RSIIndicator(close=coca['close'])
stochastic_oscilator=momentum.StochasticOscillator(close=coca['close'],high=coca['high'],low=coca['low'])
exponential_moving_avarge=trend.EMAIndicator(close=coca['close'])

```

```python
coca['accumulation']=accumulation.acc_dist_index()
coca['ADX']=avarge_directional_index.adx()
coca['MACD']=moving_avarge_covergence_divergance.macd()
coca['RSI']=relative_strength_index.rsi()
coca['Stochastic Oscillator']=stochastic_oscilator.stoch()
coca['EMA']=exponential_moving_avarge.ema_indicator()
```

```python
coca.tail()
```

* Usunicie kolumn symbol, high, open, low oraz volume, one nie bd nam ju potrzebne

```python
coca= coca.drop(['symbol','high','open','low','volume'],axis=1)
coca.tail()
```

Jeszcze raz usuwam brakujce elementy, to ze wzgldu na to, e przy pocztku danych niektre wskaniki z analizy technicznej bd miay warto null, gdy w danych nie ma wystarczajco dni do tyu

```python
coca.dropna(inplace=True)
```

# Tworzenie zbiorw na ktrych bdziemy pracowa

Nasz zbir X skada si z tablicy, przechowujcej odpowiedni ilo dni do tyu (w tym przypadku 30), w kadym z dni mamy odpowiednio cene na zamkniciu giedy, oraz wskaniki z analizy technicznej w danym dniu

```python
X=[]
y=[]

prediction_days=30

days_to_predict=10

df_numpy=coca.to_numpy()

for i in range(len(df_numpy)):
    end= i + prediction_days
    out_end= end + days_to_predict
    
    if(out_end>len(df_numpy)):
       break
    
    X.append(df_numpy[i:end, :])
    y.append(df_numpy[end:out_end,0])

       
X,y= np.array(X), np.array(y)
```

```python
X.shape
```

* Importowanie potrzebnych bilbiotek do modeli pytkich

```python
from sklearn import model_selection
import scipy.stats as stats
import sklearn.linear_model as lm
from sklearn import datasets, linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
```

Zmiana tablicy X, poniewa nie udao mi si jako input do modeli uy tablicy dwuwymiarowej

```python
X2=np.reshape(X,(753,210))
```

* Train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.2, random_state=0)
```

```python
kfold=model_selection.KFold(n_splits=5,shuffle=False)
```

# Tworzenie modeli pytkich

```python

grid1 = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), linear_model.LinearRegression()),
                    [{'polynomialfeatures__degree': [1,2,3]}],
                    cv=kfold,
                    refit=True,
                          n_jobs=10,
                         pre_dispatch=20)

grid1.fit(X_train,y_train)
grid1.best_params_
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import RandomizedSearchCV
fold=model_selection.KFold(n_splits=5,shuffle=False)
grid2= RandomizedSearchCV(make_pipeline(PolynomialFeatures(degree=2),Lasso()),
                      [{'polynomialfeatures__degree':[1,2],'lasso__alpha':[0.1,0.7,1,0.5]}],
                     cv=kfold,refit=True,
                          n_jobs=10,
                         pre_dispatch=20)
grid2.fit(X_train,y_train)
grid2.best_params_
```

```python
grid3= RandomizedSearchCV(make_pipeline(PolynomialFeatures(degree=2),ElasticNet()),
                      [{'polynomialfeatures__degree':[2,3],'elasticnet__alpha':[0.5,0.1,1,0.2,0.05]}],
                     cv=kfold,refit=True,
                          n_jobs=10,
                         pre_dispatch=20)
grid3.fit(X_train,y_train)
grid3.best_params_
```

```python
grid4= RandomizedSearchCV(make_pipeline(PolynomialFeatures(degree=2),Ridge()),
                      [{'polynomialfeatures__degree':[1,2,3,4],'ridge__alpha':[0.01,0.1,1,4,10,20]}],
                     cv=kfold,refit=True,
                          n_jobs=10,
                         pre_dispatch=20)
grid4.fit(X_train,y_train)
grid4.best_params_
```

```python
grid5= RandomizedSearchCV(RandomForestRegressor(),
                      [{'max_depth':[10,11,12,13],'n_estimators':[100,200,180,150,250]}],
                     cv=kfold,refit=True,
                          n_jobs=10,
                         pre_dispatch=20)
grid5.fit(X_train,y_train)
grid5.best_params_
```

# Sprawdzenie wynikw na tescie

```python
from sklearn import metrics

models = [] 
models.append(('LR',grid1.best_estimator_))
models.append(('Lasso',grid2.best_estimator_))
models.append(('ElasticNet',grid3.best_estimator_))
models.append(('Ridge',grid4.best_estimator_))
models.append(('RandomForest',grid5.best_estimator_))

r2 = []
explained_variance_score = []
median_absolute_error = []
mean_squared_error = []
mean_absolute_error = []
for name, model in models:
    r2.append(metrics.r2_score(y_test, model.predict(X_test)))
    explained_variance_score.append(metrics.explained_variance_score(y_test, model.predict(X_test)))
    median_absolute_error.append( metrics.median_absolute_error(y_test, model.predict(X_test)))
    mean_squared_error.append(metrics.mean_squared_error(y_test, model.predict(X_test)))
    mean_absolute_error.append(metrics.mean_absolute_error(y_test, model.predict(X_test)))

    
d = {'r2': r2, 
     'explained_variance_score': explained_variance_score, 
     'median_absolute_error': median_absolute_error,
     'mean_squared_error' : mean_squared_error,
     'mean_absolute_error' : mean_absolute_error,
    }
df = pd.DataFrame(data=d)
df.insert(loc=0, column='Method', value=['LR','Lasso','ElasticNet','Ridge','RandomForest'])
df
```

Same wyniki liczbowe nie daj nam tutaj dobrego obrazu co si w zasadzie dzieje, przy modelach tworzonych przeze mnie wczesniej wyniki R2 byy wysokie, ale na wykresie wida byo,
e modele nic nie przewiduj (wykres przewidywany by niemale tym samym co wykres realnych danych, z tym e przesunity o odpowiedni ilo dni)

# Wizualizacja wynikw modeli pytkich

Tworz funkcj prediction(model), aby otrzyma wyniki z danych testowych. Funkcja ta przechodzi po zbiorze X_test od pocztku do koca co 10 dni (poniewa model przewiduje 10 dni do przodu) i dopisuje 
swoje predykcje do tablicy predictions

```python
def prediction(model):    
    predictions=[]
    prediction=model.predict(X_test)
    for i in range(0,len(prediction),10):
        for j in range(0,10):
            predictions.append(prediction[i][j])
                
    predictions=np.array(predictions)
    return predictions
```

```python
final_y=[]
for i in range(0,len(y_test),10):
    for j in range(0,10):
        final_y.append(y_test[i][j])
final_y=np.array(final_y)
print(final_y.shape)
plt.plot(final_y)
plt.show()
```

Ptla ktra rysuje wykresy realnych danych(wykres czarny), oraz predykcji danych modeli(wykres pomaraczowy)

```python
for name,model in models:
    predicted=prediction(model)
    plt.figure(figsize=(16,8))
    plt.title(name)
    plt.plot(final_y,label='Actual Data',color='black')
    plt.plot(predicted,label=name+'Predicted',color='orange')
    plt.show
```

Sprbujmy zatem zrobi model ktry bdzie dziaa na danych przeskalowanych, oraz korzysta z dodatkowych funkcji 

```python
from sklearn.preprocessing import MinMaxScaler
```

* Skalowanie danych

```python
scaler= MinMaxScaler(feature_range=(0,1))

scaled_data=pd.DataFrame(scaler.fit_transform(coca),columns=coca.columns,index=coca.index)
```

Analogicznie jak wczeniej tworzymy zbiory X oraz y

```python
X=[]
y=[]

prediction_days=30

days_to_predict=10

df_numpy=scaled_data.to_numpy()

for i in range(len(df_numpy)):
    end= i + prediction_days
    out_end= end + days_to_predict
    
    if(out_end>len(df_numpy)):
       break
    
    X.append(df_numpy[i:end, :])
    y.append(df_numpy[end:out_end,0])

       
X,y= np.array(X), np.array(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Przygotowanie tablicy final_y aby narysowa wyniki modelu

```python
final_y=[]
for i in range(0,len(y_test),10):
    for j in range(0,10):
        final_y.append(y_test[i][j])
final_y=np.array(final_y)
print(final_y.shape)
plt.plot(final_y)
plt.show()
```

* Dodanie early_stopping

```python
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, mode='min', verbose=1)
```

* Tworzenie modelu

```python
model = Sequential()

model.add(LSTM(30,activation='tanh',return_sequences=True,input_shape=(prediction_days,7)))
model.add(Dropout(0.1))
model.add(LSTM(15,activation='tanh',return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(10,activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(days_to_predict))

model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

* uczenie modelu na przeskalowanych danych

```python
history = model.fit(X_train, y_train, validation_data= (X_test, y_test), batch_size=32,epochs=40, callbacks=[early_stopping])
```

* Rysowanie krzywej uczenia si

```python
plt.figure(figsize=(16,5))
plt.plot(history.history['loss'],color='blue')
plt.plot(history.history['val_loss'],color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
def deep_prediction(model):
    predictions=[]
    prediction=model.predict(X_test)
    for i in range(0,len(prediction),10):
        for j in range(0,10):
            predictions.append(prediction[i][j])                
    predictions=np.array(predictions)
    return predictions
```

* Wizualizacja wynikw

```python
predictions=deep_prediction(model)
plt.figure(figsize=(16,8))
plt.title("Model 1")
plt.plot(final_y,label='Actual Data',color='black')
plt.plot(predictions,label='Predicted',color='orange')
plt.show
```
