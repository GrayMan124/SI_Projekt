```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data=pd.read_csv("prices-split-adjusted.csv")
data.head()
```

```python
apple_data=data[data.symbol=='AAPL']
apple_data=apple_data[apple_data.date>"2013-01-01"]
apple_data=apple_data[apple_data.date<"2016-04-01"]
apple_data['date']=pd.to_datetime(apple_data.date)
apple_data.set_index('date',inplace=True)

plt.figure(figsize=(16,8))
plt.plot(apple_data['close'])
```

```python
import ta
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from sklearn.preprocessing import MinMaxScaler
```

```python
df=ta.add_all_ta_features(apple_data,open="open",high="high",low="low",close="close",volume="volume",fillna=True)
```

```python
df.shape
```

```python
df.drop(['open', 'high', 'low', 'volume','symbol'], axis=1, inplace=True)
```

```python
close_scaler=MinMaxScaler(feature_range=(0,1))
```

```python
df=pd.DataFrame(close_scaler.fit_transform(df),columns=df.columns,index=df.index)
```

```python
X_train,y_train=[],[]
steps_in=90
steps_out=30
df.to_numpy()
for i in range(len(df)):
    end=i+steps_in
    out_end=end+steps_out
    
    if out_end>len(df):
        break
    X_tmp,y_tmp=df[i:end],df[end:out_end]['close']
    X_train.append(X_tmp)
    y_train.append(y_tmp)

X_train,y_train=np.array(X_train),np.array(y_train)    
```

```python
print(X_train[1].shape)
print(y_train[1].shape)
```

```python
def validater(n_per_in, n_per_out,n_features):
    """
    Runs a 'For' loop to iterate through the length of the DF and create predicted values for every stated interval
    Returns a DF containing the predicted values for the model with the corresponding index values based on a business day frequency
    """
    
    # Creating an empty DF to store the predictions
    predictions = pd.DataFrame(index=df.index, columns=[df.columns[0]])

    for i in range(n_per_in, len(df)-n_per_in, n_per_out):
        # Creating rolling intervals to predict off of
        x = df[-i - n_per_in:-i]

        # Predicting using rolling intervals
        yhat = model.predict(np.array(x).reshape(1, n_per_in, n_features))

        # Transforming values back to their normal prices
        yhat = close_scaler.inverse_transform(yhat)

        # DF to store the values and append later, frequency uses business days
        pred_df = pd.DataFrame(yhat, 
                               index=pd.date_range(start=x.index[-1], 
                                                   periods=len(yhat), 
                                                   freq="B"),
                               columns=[x.columns[0]])

        # Updating the predictions DF
        predictions.update(pred_df)
        
    return predictions
```

```python
model=Sequential()

model.add(LSTM(90,activation='tanh',return_sequences=True, input_shape=(90,df.shape[1])))
model.add(Dropout(0.2))
model.add(LSTM(40, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(60))
model.add(Dropout(0.2))
model.add(Dense(30))

model.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X_train,y_train,epochs=50,batch_size=128,validation_split=0.1)
```

```python
def visualize_training_results(results):
    """
    Plots the loss and accuracy for the training and testing data
    """
    history = results.history
    plt.figure(figsize=(16,5))
    plt.plot(history['val_loss'])
    plt.plot(history['loss'])
    plt.legend(['val_loss', 'loss'])
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.show()
    
    plt.figure(figsize=(16,5))
    plt.plot(history['val_accuracy'])
    plt.plot(history['accuracy'])
    plt.legend(['val_accuracy', 'accuracy'])
    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.show()
    
```

```python
visualize_training_results(history)
```
