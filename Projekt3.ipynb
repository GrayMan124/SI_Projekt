```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data=pd.read_csv("prices-split-adjusted.csv")
data.head()
```

```python
apple_data=data[data.symbol=='AAPL']
apple_data=apple_data[apple_data.date>"2012-01-01"]
apple_data=apple_data[apple_data.date<"2016-04-01"]
apple_data['date']=pd.to_datetime(apple_data.date)
apple_data.set_index('date',inplace=True)

plt.figure(figsize=(16,8))
plt.plot(apple_data['close'])
```

```python
import ta
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from sklearn.preprocessing import MinMaxScaler
```

```python
apple_close=apple_data['close']
```

```python
print(apple_close)
```

```python
prediction_days=90

scaler = MinMaxScaler(feature_range=(0,1))
data1=apple_data
scaled_data_apple = scaler.fit_transform(data1['close'].values.reshape(-1,1))

x_train=[]
y_train=[]
for x in range(prediction_days,len(scaled_data_apple)-30):
    x_train.append(scaled_data_apple[x-prediction_days:x,0])
    y_train.append(scaled_data_apple[x:x+30,0])
x_train=np.array(x_train)
y_train=np.array(y_train)
x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
```

```python
model=Sequential()

model.add(LSTM(units=90, activation='tanh',return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.4))
model.add(LSTM(units=30, return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(units=60))
model.add(Dropout(0.4))
model.add(Dense(units=30))

model.compile(optimizer='adam',loss='mean_squared_error')
model.summary()
```

```python
history=model.fit(x_train,y_train,epochs=25,batch_size=128)
```

```python

plt.figure(figsize=(16,5))
plt.plot(history.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
intel_data=data[data.symbol=='INTC']
intel_data=intel_data[intel_data.date>'2015-01-01']
intel_close=intel_data["close"].values.reshape(-1,1)
plt.plot(intel_close)
```

```python
data2=intel_data
scaled_data_intel = scaler.fit_transform(data2['close'].values.reshape(-1,1))
x_test_intel=[]
for x in range(prediction_days,len(scaled_data_intel)):
    x_test_intel.append(scaled_data_intel[x-prediction_days:x,0])
    
x_test_intel=np.array(x_test_intel)
x_test_intel = np.reshape(x_test_intel,(x_test_intel.shape[0],x_test_intel.shape[1],1))


predicted_intel=model.predict(x_test_intel)
predicted_intel=scaler.inverse_transform(predicted_intel)
final_predicted_intel=[]
for i in range(0,len(predicted_intel),30):
    for j in range(30):
        final_predicted_intel.append(predicted_intel[i][j])
```

```python
plt.figure(figsize=(16,8))
plt.title("Intel")
plt.plot(intel_close[prediction_days:],label='Intel Actual Data',color='black')
plt.plot(final_predicted_intel[6:],label='Predicted Intel',color='orange')
plt.show
```
